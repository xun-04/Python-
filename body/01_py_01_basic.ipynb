## åˆ—è¡¨æ¨å¯¼

åœ¨Pythonä¸­ï¼Œå¯ä»¥ç»“åˆå¾ªç¯è¯­å¥forã€åˆ¤æ–­è¯­å¥ifã€elseç­‰ï¼Œå†™å‡ºæ›´åŠ ç®€æ´çš„ç¨‹åºã€‚

ä»¥ä¸Šä»£ç å·²ç»è¶³å¤Ÿç®€æ´ï¼Œä½†æ˜¯Pythonè¿˜æ”¯æŒä»¥ä¸‹æ›´ç®€æ´çš„æ–¹å¼ï¼š

````python
a=[2*i+1 for i in range(50) if (2*i+1)%3==0]
print(a)
````

````python
b=[(2*i+1,2*j+1) for i in range(50) for j in range(50) if (2*i+1)%3==0 and (2*j+1)%5==0]
print(b)
````

## å°ç»“

åœ¨ä»¥ä¸Šçš„ç¨‹åºä¸­ï¼Œæˆ‘ä»¬é¦–å…ˆä½¿ç”¨input()å‡½æ•°è·å¾—ç”¨æˆ·è¾“å…¥ï¼Œæ¥ç€åˆ¤æ–­ç”¨æˆ·è¾“å…¥çš„æ˜¯å¦ä¸º"end"ï¼Œå¦‚æœæ˜¯ï¼Œåˆ™é€€å‡ºï¼Œå¦åˆ™ï¼Œç”±äºinput()è·å¾—çš„æ˜¯ä¸€ä¸ªå­—ç¬¦ä¸²ï¼Œå› è€Œæˆ‘ä»¬ä½¿ç”¨float()å‡½æ•°å°†è¯¥å­—ç¬¦ä¸²è½¬æ¢ä¸ºä¸€ä¸ªæµ®ç‚¹æ•°å­—ï¼Œå†å°†å…¶ç›¸åŠ ã€‚

ä½†æ˜¯ä»¥ä¸Šç¨‹åºå¹¶ä¸å®Œç¾ï¼Œæ¯”å¦‚ï¼Œå¦‚æœç”¨æˆ·è¾“å…¥çš„ä¸æ˜¯â€œendâ€ä¹Ÿä¸æ˜¯æ•°å­—ï¼Œå°±ä¼šæŠ¥é”™ï¼Œå› è€Œä¸€ä¸ªæ›´å®Œå–„çš„ç‰ˆæœ¬æ˜¯åŠ å…¥é”™è¯¯å¤„ç†ï¼š

````python
summ=0
while True:
    text=input("è¯·è¾“å…¥ä¸€ä¸ªæ•°å­—ï¼Œæˆ–è€…endç»“æŸï¼š")
    if text=="end":
        break
    try:
        summ+=float(text)
    except Exception as e:
        print("è¯·è¾“å…¥æ•°å­—æˆ–endï¼")
        continue
print(summ)
````
